```python
from tools.stream import StreamCCTV

import cv2
from matplotlib import pyplot as plt
```

```python
cam = StreamCCTV(
                url="rtsp://210.99.70.120:1935/live/cctv001.stream",
                transport="tcp",
                max_width=1920
                ).start()
cam.wait_ready(2.0)
```

```python
frame = cam.shoot(wait=True, timeout=2.0)  # 최신 1장
if frame is None:
    print("프레임 실패")
else:
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)); plt.axis('off'); plt.show()
```

```python
cam.stop()
```
